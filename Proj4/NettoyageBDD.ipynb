```python
transactions.loc[transactions["session_id"]=="s_0"]
#Une courte recherche me permet de trouver les 200 lignes coupables
#Plusieurs recherches permettent de vérifier que l'id_prod T_0 et les clients ct_1 et ct_0
#ne sont dans aucune autre ligne de la table transactions
```

```python
transactionsclean=transactions.loc[transactions["id_prod"]!="T_0"]
```

```python
produitsclean=produits.loc[produits["price"]>=0]
```

```python
clientsclean=clients[clients.client_id!="ct_1"]
```

```python
clientsclean=clientsclean[clientsclean.client_id!="ct_0"]
```

```python
#df.info() m'a permis de vérifier qu'aucune valeur n'était manquante jusqu'ici
#A ce stade les 3 df sont "clean" : j'ai retiré 200 lignes dans les transactions.
#Ces lignes correspondaient à des tests, et les utilisateurs ct_0 et ct_1
#ainsi que le produit T_0 n'ont été créé qu'à cet effet.
#C'est pourquoi nous les retirons aussi dans les tables clients et produits.
```

```python
transactions_prix=pd.merge(transactionsclean,produitsclean,on="id_prod",how="left")
#left join par sécurité
```

```python
transactions_prix.info()
#Nous constatons tout de suite un problème: price et categ contiennent chacun 103 lignes nulles
```

```python
transactions_prix[transactions_prix.isnull().any(axis=1)]
#Nous avons trouvé le produit 0_2245 qui n'a pas de référence dans le tableau produits
#Cependant le premier chiffre 0 nous indique la catégorie a laquelle il appartient.
```

```python
produitsclean.groupby(by="categ").mean()
```

```python
transactions_prix["categ"] = transactions_prix[["categ"]].fillna(value=0)
transactions_prix["price"] = transactions_prix[["price"]].fillna(value=11.732795)
#Après avoir calculé le prix moyen des produits selon la catégorie,
#je décide de faire une imputation par la moyenne à ce produit.
```

```python
transactions_full=pd.merge(transactions_prix,clientsclean, on="client_id",how="left")
#Dernier fusion de table
```

```python
transactions_full.info()
```
